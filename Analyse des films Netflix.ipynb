## Analyse de l'ensemble des données disponibles sur Netflix  jusqu'en 2021

## Présenetation des données

- **show_id** correspond à l'ID unique du film ou de la série
- **type** permet de dire s'il s'agit d'un film ("Movie") ou d'une série télévisée ("TV show")
- **title** le tire du film/de la série
- **director** le directeur de production
- **cast** les acteurs présent dans le film/la série
- **country** pays dans lequel le film/la série à été produite
- **date_added** date à laquelle le film/ la série a été ajouté sur netflix
- **release_year** l'année de sortie réelle du film/de la série
- **rating** classement du contenu
- **duration** pour les films durée en munites/ pour les séries nombre de saisons
- **listed_in** la ou les catégories du film/de la série (drame, comédie, documentaire, etc.)
- **description** la description du film/de la série

## Afficher l'évolution du nombre de film/séries disponible sur Netflix au cours du temps

```python
donnees_film = pd.read_csv(path + 'netflix_titles.csv', index_col = [0], parse_dates = [6])
donnees_film.head(2)
```

```python
from datetime import datetime
donnees_film_NaN = donnees_film[donnees_film['date_added'].notna()]
##donnee_film["date_added"] = pd.to_datetime(donnee_film["date_added"], infer_datetime_format=True)
```

```python
donnees_film_NaN["year_added"] = donnees_film_NaN['release_year'].astype('int')
```

```python
donnees_film_NaN.info()
```

```python
donnees_film_NaN.groupby(['year_added', 'type']).size().reset_index(name = "count").head()
```

```python
plt.figure(figsize = [17, 6])

donnees_pointplot = donnees_film_NaN.groupby(["year_added", "type"]).size().reset_index(name = "count")
sns.pointplot( x = "year_added", y = "count", data = donnees_pointplot, hue = "type")
plt.xticks(rotation = 90)
plt.show()
```

```python
series_10_pays = donnees_film_NaN['country'].value_counts().head(10)
donnee_pays = donnees_film_NaN[donnees_film_NaN['country'].isin(series_10_pays.keys())]
donnees_point = donnee_pays.groupby(["year_added", "country"]).size().reset_index(name = "count")
plt.figure(figsize = [15,6])
sns.pointplot(x = "year_added", y = "count", data = donnees_point, hue = "country")
plt.xticks(rotation = 90)
plt.show()
```
